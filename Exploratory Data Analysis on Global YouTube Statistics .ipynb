## Loading the Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Global YouTube Statistics.csv", encoding='latin1')
```

```python
category_subscribers_mean = df.groupby("category")["subscribers"].mean()
category_subscribers_mean
```

```python
category_subscribers_mean.sort_values(ascending = False,inplace=True)
```

```python
plt.bar(category_subscribers_mean.index,category_subscribers_mean.values)
plt.title("Mean Subscribers by Category")
plt.xlabel("Category")
plt.ylabel("Mean Subscribers")
plt.xticks(rotation=90)
plt.show()
```

```python
channel_type_counts = df["channel_type"].value_counts()
channel_type_counts
```

```python
channel_type_counts.plot(kind="pie")
plt.title("Channel Type Distribution")
plt.show()
```

```python
sns.scatterplot(x="subscribers",y="video views",data=df)
plt.title("Scatter Plot of Subscribers vs Video Views")
plt.xlabel("Subscribers")
plt.ylabel("Video Views")
plt.show()
```

```python
country_subscribers_total = df.groupby("Country")["subscribers"].sum().sort_values(ascending=False).head(20)
country_subscribers_total
```

```python
plt.bar(country_subscribers_total.index,country_subscribers_total.values)
plt.title("Top 20 Countries with highest number of subscribers")
plt.xlabel("country")
plt.ylabel("Number of Subscribers")
plt.xticks(rotation=90)
plt.show()
```
